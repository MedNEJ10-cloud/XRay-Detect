```python
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from imutils import paths
import numpy as np
import os
import cv2
```

```python

imagePaths = list(paths.list_images(r"D:\BTS\2DAI\PFE\Cours\covid19_dataset"))
data_train = []
labels_train = []
for imagePath in imagePaths:
    label = imagePath.split(os.path.sep)[-2]
    image = cv2.imread(imagePath)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (224, 224))
    data_train.append(image)
    labels_train.append(label)
```

```python
#nomrilzation
np.array(data_train) / 255.0
#encoding
lb = LabelBinarizer()
labels_train = lb.fit_transform(labels_train)
```

```python
(trainX, testX, trainY, testY) = train_test_split(data_train, labels_train,
    test_size=0.20, stratify=labels_train, random_state=42)
```

```python
np.save('trainX.npy', trainX)
np.save('trainY.npy', trainY)
np.save('testX.npy', testX)
np.save('testY.npy', testY)
```
